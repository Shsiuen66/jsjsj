```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

##Question 1
plot a line plot between a and b:\
a=np.arange(40,50)\
b=np.arange(50,60)


```python
a = np.arange(40,50)
b = np.arange(50,60)
plt.plot(a,b)
plt.title('Question 1')
plt.xlabel('a values')
plt.ylabel('b values')
plt.show()
```

##Question 2
Plot a line plot showing the sales trend in company 1 and 2:\
\
days = [1,2,3,4,5,6,7] #days of d week\
sales_1 = [160,150,140,145,175,165,180] #sales of company1\
sales_2 = [70,90,160,150,140,145,175]  #sales of company2


```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2
```

```python
plt.plot(days,sales_1,color = 'blue')
plt.plot(days,sales_2,color = 'yellow')
plt.title('Sales trend ',fontsize = 10,color = 'red')
plt.ylabel('Sales in USD')
plt.xlabel('Day of the week')
plt.show()
```

##Question 3
Create a 3 by 3 subplots:\
multiple plots\
x = [1,2,3,4]\
y1 = [4,3,2,1]\
y2 = [10,20,30,40]\
y3 = [40,30,20,10]\
y4 = [1,2,1,2]\
y5 = [40,70,90,70]


```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.figure(figsize = (8,8))
plt.subplot(3,3,1)
plt.plot(x,y1,label='y1',color='red',marker='.',markersize = 6)
plt.title('SubPlot 1')
plt.legend(shadow=True)

plt.subplot(3,3,2)
plt.plot(x,y2,label='y2',color='blue',marker='.',markersize = 6)
plt.title('SubPlot 2')
plt.legend(shadow=True)

plt.subplot(3,3,3)
plt.plot(x,y3,color='green',label='y3',marker='.',markersize = 6)
plt.title('SubPlot 3')
plt.legend(shadow=True)

plt.subplot(3,3,4)
plt.plot(x,y4,color='yellow',label='y4',marker='.',markersize = 6)
plt.title('SubPlot 4')
plt.legend(shadow=True)

plt.subplot(3,3,5)
plt.plot(x,y5,color='pink',label='y5',marker='.',markersize = 6)
plt.title('SubPlot 5')
plt.legend(shadow=True)

plt.show()
```
